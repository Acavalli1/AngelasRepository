### Insight 1) 

##### More men than any other gender, are taking computing in order to get a job in the computing field. The pie chart below shows how Men are over represented in "to get a job in computing." 

###### The data used was from new_df.groupby('Gender').count()

```python
labels = 'Non Binary', 'does not identify', 'Man', 'Prefers not to say', 'Woman'
sizes = [3, 1, 120, 5, 31]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
plt.title('Men Seek Computing Degrees, to Get a Job in Computing, Overshadowing All Other Gender Categories')
```

### Insight 2) 

##### Of the students who are seeking computing degrees to get a job in computing, what is their most represented age group? This is important data that answers the question, how old are the students when they are working towards a degree in computing? The bar graph below shows that most students who are "motivated for a job in computing," are in the 18 and under OR 19-20 age groups. 

###### The data on the bar graph below, shows that most of the students are very young --mostly in high school--when they begin computing courses, and working towards a degree. Why are more students at the high school level, active in the field, vs. the other age categories? Recruitment can occur in areas where these age groups organically exist (e.g., high schools).

##### The data used was from: new_df.groupby('Age ').count()

```python
df = pd.DataFrame({'lab':['18 and under', '19-20', '21-24', '25-34', '35-64', '65 +'], 'val':[58, 52, 32, 10, 7, 1]})
ax = df.plot.bar(x='lab', y='val', rot=0)
plt.title('Most Students -- Motivated for a Job in Computing -- are in the 18 and Under or 19-20 Age Groups')
```

### Insight 3)

##### The majority of students who are "motivated for a job in computing" are White/Caucasian. If looking to recruit those under represented in race/ethnicity, you can use the bar graph below.

##### The data used was from: new_df.groupby('Race/ethnicity').count()   -- to get the values for each race/ethnicity. 

```python
df = pd.DataFrame({'lab':['Asian', 'His/Lat', 'His;W/C', 'multRac', 'W/C', 'B/AA', 'no reply'], 'val':[29, 30, 14, 3, 64, 3, 7]})
ax = df.plot.bar(x='lab', y='val', rot=0)
plt.title('Most Students -- Motivated for a Job in Computing -- are White/Caucasian, African Americans are Underrepresented')
```

### Insight 4) 

#### From the visualization below, you can see that nearly half of the students in the IT Certificate program, are students who changed from another degree program.
